---------- Creting an new Columns --------------

```python
df["result"]=df["report_card\\"].apply(lambda x : "pass" if x >= 50 else "fail")
```

```python
print(df.result).head(3)
```

```python
print(df["result"].head(3))
```

```python
# count of pass and fail
print(df["result"].value_counts())

# percentage of pass and fail
print(df["result"].value_counts(normalize=True).round(2) * 100)

# average report_card\\ score of pass students
print(df[df["result"] == "pass"]["report_card\\"].mean().round(2))

# average report_card\\ score of fail students
print(df[df["result"] == "fail"]["report_card\\"].mean().round(2))
```

```python
# find Q1 and Q3
Q1 = df["report_card\\"].quantile(0.25)
Q3 = df["report_card\\"].quantile(0.75)
IQR = Q3 - Q1

# filter out outliers
df = df[(df["report_card\\"] >= Q1 - 1.5*IQR) &
        (df["report_card\\"] <= Q3 + 1.5*IQR)]
```

```python
# step 1 — check column name exists
print(df.columns)

# step 2 — check Q1 Q3 IQR values
print("Q1  :", Q1)
print("Q3  :", Q3)
print("IQR :", IQR)

# step 3 — check rows before and after outlier removal
print("Total rows:", df.shape[0])

# step 4 — check min and max of report_card\\
print("Min score:", df["report_card\\"].min())
print("Max score:", df["report_card\\\\"].max())
```
